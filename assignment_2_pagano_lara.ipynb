- Diagonal Difference

```python
#!/bin/python3

import math
import os
import random
import re
import sys

def diagonalDifference(arr):
    # Write your code here
    D1 = 0
    D2 = 0
    l= len(arr)
    for i in range(0, l):
        D1 += arr[i][i]
        D2 += arr[i][l-(i+1)]
    diff= D1-D2
    if diff < 0:
        diff = diff * -1
    return diff 
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = []

    for _ in range(n):
        arr.append(list(map(int, input().rstrip().split())))

    result = diagonalDifference(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

- Running Time of Algorithms

```python
#!/bin/python3

import math
import os
import random
import re
import sys

def runningTime(arr):
    # Write your code here
    count = 0
    for i in range(1, len(arr)):
        j = i-1
        pos = arr[i]
        while (j >= 0) and (arr[j] > pos):
            arr[j+1] = arr[j]
            j -= 1
            count += 1
            arr[j+1] = pos
    return count
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = runningTime(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

- Strong Password

```python
import math
import os
import random
import re
import sys

def minimumNumber(n, password):
    # Return the minimum number of characters to make the password strong
    special_characters = "!@#$%^&*()-+"
    d= 0
    l= 0
    u= 0
    t= 0

    for i in password:
        if i.isdigit():
            d = 1
        elif i.islower():
            l = 1
        elif i.isupper():
            u = 1
        elif i in special_characters:
            t = 1   
    
    needed_ch= d+ l+u+ t
    len_pss= 6 - n
    missing= 4 - needed_ch
    return max( len_pss , missing) 

   
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    password = input()

    answer = minimumNumber(n, password)

    fptr.write(str(answer) + '\n')

    fptr.close()



if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    password = input()

    answer = minimumNumber(n, password)

    fptr.write(str(answer) + '\n')

    fptr.close()
```

- Super Reduced String

```python
import math
import os
import random
import re
import sys

def superReducedString(s):
    # Write your code here
    for i in range(0,len(s)-1):
            if s[i] == s[i+1]:
                return superReducedString(s[:i]+s[i+2:])   
    else:        
            if len(s) == 0: 
                return "Empty String"
            return s
                
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = superReducedString(s)

    fptr.write(result + '\n')

    fptr.close()
```
