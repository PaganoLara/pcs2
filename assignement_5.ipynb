## mprt   

```python
import requests
import re

link = 'https://www.uniprot.org/uniprot/'
 
with open("rosalind_mprt (6).txt") as file:
    seq = file.read().replace("\n", " ").split()
sequences = {}
for seq_id in seq:
    old_seq_name = seq_id[0:6]
    goToURL = link + old_seq_name + ".fasta"
    response = requests.get(goToURL)
    sequences[seq_id] = (response.text.split("\n"))
    sequences[seq_id] = "".join(sequences[seq_id][1::])

keys= sequences.keys() 
for i in keys:
    pos = []
    p = re.compile(r'N[^P](S|T)[^P]')
    for j in range(len(sequences[i])-3):
        if p.match(sequences[i][j:j+4]):
            pos.append(str(j+1))
    if len(pos) > 0:
        print(i)
        print(' '.join(pos))
```

## orf

```python
import re                                                 
from Bio import SeqIO                                     
from Bio.Seq import Seq                                   

f = SeqIO.read('rosalind_orf (1).txt', 'fasta')  
Orfs = re.compile(r'(?=(ATG(?:...)*?)(?=TAG|TGA|TAA))')
seq1 = f.seq                             
seq_rv = seq1.reverse_complement()    
st_seq1= str(seq1)
st_seq_rv= str(seq_rv)

results = []                                            
for m in re.findall(Orfs, st_seq1):               
    dna = Seq(m)                
    protein = dna.translate()                        
    if protein not in results:                         
        results.append(protein)  
                  
for n in re.findall(Orfs, st_seq_rv):               
    dna_rev = Seq(n)            
    rev_protein = dna_rev.translate()                
    if rev_protein not in results:                     
        results.append(rev_protein)                    

for i, s in enumerate(results):                         
    print(s)                     
```

## SPLC	RNA Splicing

```python

from Bio import SeqIO

coding_table = {'TCT': 'S',   'TCC': 'S',   'TCA': 'S',   'TCG': 'S', 
                'TTT': 'F',   'TTC': 'F',   'TTA': 'L',   'TTG': 'L',
                'CTT': 'L',   'CTC': 'L',   'CTA': 'L',   'CTG': 'L',
                'TAT': 'Y',   'TAC': 'Y',   'TGT': 'C',   'TGC': 'C',
                'TGG': 'W',   'CCT': 'P',   'CCC': 'P',   'CCA': 'P',
                'CCG': 'P',   'CAT': 'H',   'CAC': 'H',   'CAA': 'Q', 
                'CAG': 'Q',   'CGT': 'R',   'CGC': 'R',   'CGA': 'R',
                'CGG': 'R',   'AGA': 'R',   'AGG': 'R',   'ATT': 'I',
                'ATC': 'I',   'ATA': 'I',   'ATG': 'M',   'ACT': 'T',
                'ACC': 'T',   'ACA': 'T',   'ACG': 'T',   'AAT': 'N',
                'AAC': 'N',   'AAA': 'K',   'AAG': 'K',   'GTT': 'V',
                'GTC': 'V',   'GTA': 'V',   'GTG': 'V',   'GCT': 'A',
                'GCC': 'A',   'GCA': 'A',   'GCG': 'A',   'GAT': 'D',
                'GAC': 'D',   'GAA': 'E',   'GAG': 'E',   'GGT': 'G', 
                'GGC': 'G',   'GGA': 'G',   'GGG': 'G',   'AGT': 'S',
                'AGC': 'S',   'TAA': '*',   'TAG': '*',   'TGA': '*'}


def RNA_splicing(dna_string, introns):
    for i in introns:
        dna_string = dna_string.replace(i, "")
    
    protein_string  = ""
    for i in range(0, len(dna_string)-2, 3):
        if coding_table[dna_string[i:i+3]] == "*":
            break
        protein_string += coding_table[dna_string[i:i+3]]
    return protein_string

seq_id= []
seq = []
with open ("rosalind_splc (1).txt",'r') as f:
    for x  in SeqIO.parse(f,'fasta'):
        seq_id.append(str(x.name))
        seq.append(str(x.seq))

dna= seq[0]
intron = seq[1:]
rna = RNA_splicing(dna, intron)

print(rna)
```

## pmch

```python

from Bio import SeqIO
from math import factorial

def PerfectMatch(s):
    a = s.count('A')
    c = s.count('C')    
    return factorial(a) * factorial(c) 

seq_id= []
seq= []
with open ("rosalind_pmch.txt",'r') as f:
    for x  in SeqIO.parse(f,'fasta'):
        seq_id.append(str(x.name))
        seq.append(str(x.seq))

final = seq[0]

print(PerfectMatch(final))

```

## pper 

```python
from math import factorial

def PartialPermutations(n, k):
	return factorial(n)/factorial(n-k) % 10**6

f= open("rosalind_pper.txt","r")
line= f.readline().split()
num= [int(x) for x in line]
n= num[0]
k= num[1]
print(int(PartialPermutations(n, k)))


```

## long

```python
from Bio import SeqIO 
def SSuperstring(rds, sprS=''):
    if len(rds) == 0:
        return sprS

    elif len(sprS) == 0:
        sprS = rds.pop(0)
        return SSuperstring(rds, sprS)

    else:
        for i in range(len(rds)):
            r = rds[i]
            lEn = len(r)

            for t in range(lEn // 2):
                ovr_len = lEn - t

                if sprS.startswith(r[t:]):
                    rds.pop(i)
                    return SSuperstring(rds, r[:t] + sprS)

                if sprS.endswith(r[:ovr_len]):
                    rds.pop(i)
                    return SSuperstring(rds, sprS + r[ovr_len:])

seq = []
with open ("rosalind_long (1).txt",'r') as f:
    for x  in SeqIO.parse(f,'fasta'):
        seq.append(str(x.seq))

    print(SSuperstring(seq))
```

## tran

```python
from Bio import SeqIO   

seq = []
with open ("rosalind_tran.txt",'r') as f:
    for x  in SeqIO.parse(f,'fasta'):
        seq.append(str(x.seq))
s1 = seq[0]
s2 = seq[1]

def Tran(s1, s2):
		t=[0,0]
		nucleotides= [['A', 'G'],['C', 'T']]
		pyrimidine = ['C','T']
		for i in range(len(s1)):
			if s1[i] == s2[i]:
				pass
			elif s1[i] in nucleotides[s2[i] in pyrimidine]:
				t[0] += 1
			else:
				t[1] +=1
		if t[1] != 0:
        		return round((t[0] / t[1]),11)

Tran(s1, s2)
```
